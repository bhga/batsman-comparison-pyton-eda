Who is better KL Rahul or Shubnam Gill??

As cricket is played in three time format taking only Run Avg should not be the only criteria for comparison 

Also thinking about the players position player as in our comparison both players are opening batsman

As this players are more likely of playing under power-play which is crucial for the team it is must that the batting team should take advantage of the feild set 

Hence for comparison we are looking for player who has hard hitting ability and scores the run fast

```python
import pandas as pd
```

```python
kl_rahul_df = pd.read_excel('/content/sample_data/KL_Rahul.xlsx')
```

```python
kl_rahul_df
```

```python
shb_gill_df = pd.read_excel('/content/sample_data/Shubman_Gill.xlsx')
```

```python
kl_rahul_df_test = kl_rahul_df.loc[(kl_rahul_df['Format'] == 'Test')]
```

```python
kl_rahul_df_test
```

```python
kl_rahul_df_test.loc[:,"HardHitting"]=(4*kl_rahul_df_test["4s"]+6*kl_rahul_df_test["6s"])/kl_rahul_df_test["BF"]
kl_rahul_df_test.loc[:,"FastScoring"]=kl_rahul_df_test["Runs"]/kl_rahul_df_test["BF"]
kl_rahul_df_test.loc[:,"Running"]=(kl_rahul_df_test["Runs"]-(4*kl_rahul_df_test["4s"]+6*kl_rahul_df_test["6s"]))/(kl_rahul_df_test["BF"]-(kl_rahul_df_test["4s"]+kl_rahul_df_test["6s"]))
```
